# Final Project: Spotify Top Songs Analysis

## Exploratory Analysis

```python
# Import dependencies
import pandas as pd
import psycopg2
import numpy as np
```

### Load in data from database

```python
from sqlalchemy import create_engine
from getpass import getpass

# Set up connection to database
password = getpass('Enter database password.')
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/project_spotify_db')
```

```python
# Read in Spotify song data from database
song_df = pd.read_sql('SELECT * FROM top_songs;', engine, index_col='Index')
song_df.head()
```

### Song Analysis

#### What songs have charted the most?

```python
song_number_charted_df = song_df[['song_id', 'song_name', 'artist', 'number_of_times_charted']]
song_number_charted_df = song_number_charted_df.sort_values(['number_of_times_charted'], ascending=False)
song_number_charted_df.head(10)
```

Most charted song since Jan. 2020: 'Falling' by Harry Styles

#### What songs have the highest number of streams?

```python
song_streams_df = song_df[['song_id', 'song_name', 'artist', 'streams']]
song_streams_df = song_streams_df.sort_values(['streams'], ascending=False)
song_streams_df.head(10)
```

Song with most streams since Jan. 2020: 'Beggin'' by Måneskin

### Artist Analysis

```python
# Create copy of dataframe focusing on artist info
artist_df = song_df[['song_id', 'song_name', 'artist', 'number_of_times_charted', 'highest_charting_position', 'number_weeks_charted', 'streams', 'artist_followers']]
artist_df.head()
```

#### Which artist has the most charted songs?

```python
artist_charted_grouped = artist_df[['artist', 'song_name']].groupby(['artist'])

# Convert GroupBy object into DataFrame
artist_songs_count_df = pd.DataFrame(artist_charted_grouped['song_name'].count())
artist_songs_count_df = artist_songs_count_df.rename(columns={'song_name':'count_of_charted_songs'})
artist_songs_count_df = artist_songs_count_df.sort_values(['count_of_charted_songs'], ascending=False)
artist_songs_count_df.head(10)
```

Artist with most charted songs since Jan. 2020: Taylor Swift

#### Which artist has the most streams?

```python
artist_streams_grouped = artist_df[['artist', 'streams']].groupby(['artist'])

# Convert GroupBy object into DataFrame
artist_streams_total_df = pd.DataFrame(artist_streams_grouped['streams'].sum())
artist_streams_total_df = artist_streams_total_df.sort_values(['streams'], ascending=False)
artist_streams_total_df.head(10)
```

Artist with the most streams since Jan. 2020: Taylor Swift

#### Which artists have the most #1 hits?

```python
# Create dataframe of number one songs only
artist_hits_df = artist_df.loc[artist_df["highest_charting_position"] == 1, ['artist', 'song_name', 'highest_charting_position']]
artist_hits_df.head()
```

```python
artist_hits_grouped = artist_hits_df[['artist', 'song_name']].groupby(['artist'])

# Convert GroupBy object into DataFrame
total_hits_by_artist_df = pd.DataFrame(artist_hits_grouped['song_name'].count())
total_hits_by_artist_df = total_hits_by_artist_df.rename(columns={'song_name':'count_of_number_ones'})
total_hits_by_artist_df = total_hits_by_artist_df.sort_values(['count_of_number_ones'], ascending=False)
total_hits_by_artist_df.head(10)
```

Artist with most #1 hits since Jan. 2020: Olivia Rodrigo

#### Which artists have the most followers?

```python
artist_followers_grouped = artist_df[['artist', 'artist_followers']].groupby(['artist'])

# Convert GroupBy object into DataFrame
artist_followers_df = pd.DataFrame(artist_followers_grouped['artist_followers'].sum())
artist_followers_df = artist_followers_df.sort_values(['artist_followers'], ascending=False)
artist_followers_df.head(10)
```

Artist with the most followers: Taylor Swift
